```python
from pygooglenews import GoogleNews
import pandas as pd
```

```python
gn = GoogleNews(lang='en',country="US")
```

```python
search = gn.search('Nasdaq')
```

```python
for i in search['entries']:
  print(i)
```

```python
def get_titles(keyword):
  news= []
  gn=GoogleNews(lang='en',country='US')
  search = gn.search(keyword)
  articles = search['entries']
  for i in articles:
   article= {'title': i.title, 'link': i.link,"published":i.published}
   news.append(article)
  return news

data = get_titles("Nasdaq")
data
```

```python
df = pd.DataFrame(data)
```

```python
from textblob import TextBlob
```

```python
blob = TextBlob('Apple')
```

```python
def sentiment(text):
  blob=TextBlob(text)
  return blob.sentiment.polarity

df['sentiment'] =df['title'].apply(sentiment)
```

```python
df
```

```python
import numpy as np

df['Sentiment Class']  = np.where(df['sentiment']<0,"negative",
                                  np.where(df['sentiment']>0,"positive",
                                           "neutral"))
df
```

```python
df['Sentiment Class'].value_counts(normalize=True)
```

```python
df['Date'] = pd.to_datetime(df['published'])
df['Date'] =df['Date'].dt.date
```

```python
df = df.sort_values(by='Date', ascending=False)
df
```
